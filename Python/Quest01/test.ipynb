```python
# 단어 입력 받기
word =input("입력값: \n")

# 입력 받은 단어 회문으로 만들기
word2 = "".join(reversed(word)) # word[::-1]도 실행 가능

# 제어문 사용
if word == word2:
  print("\n출력값: ")
  print(f'뒤집힌 단어는: {word} 입니다')
  print('입력된 단어는 회문입니다.')
```
